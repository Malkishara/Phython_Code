# Tkinter

This is a simple registration form for online course based on programming language.

By this form we can get following details of the student who would like to follw this courses.
1.First Name
2.Last Name
3.Gender
4.Phone number
5.Email address
6.Address
7.Department
8.What are the courses they like to follow?

After submiting that details those are be recoded in StudentDetails.txt file.

```python
import re
from tkinter import *
import tkinter as tk
regex = '^\w([\.-]?\w+)*@\w([\.-]?\w+)*(\.\w{2,3})+$'

def filedreq():
    if Fname.get() == "":
        print("First Name Field is Empty!!")
        user = "First Name Field is Empty!!"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)


    elif Lname.get() == "":
        print("Last Name Field is Empty!!")
        user = "Last Name Field is Empty!!"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)
        
    elif Phonenum.get() == "":
        print("Phone number Field is Empty!!")
        user = "Phone number Field is Empty!!"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)   

    elif Email.get() == "":
        print("Email Field is Empty!!")
        user = "Email Field is Empty!!"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)


    elif entry_Address.compare("end-1c", "==", "1.0"):    
        print("Address Field is Empty!!")
        user = "Address Field is Empty!!"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)

    else:
        checkemail()

def checkemail():
    email = Email.get()

    if(re.search(regex,email)):
        print("Valid Email")
        user = "Valid Email!!"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)
        save()

    else:
        print("Invalid Email!!")
        user = "Invalid Email-Type!! Type Correct Mail: Format: abc@x.com"
        Label(win,text=user,fg="red",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)

def save():
    First = Fname.get()
    Last = Lname.get()
    gender = str(radio.get()) 
    phone=Phonenum.get()
    email = Email.get()
    address = entry_Address.get(1.0,END)
    c = check.get()
    php = check2.get()
    java = check3.get()
    py = check4.get()
    dep = variable.get()

    if c =='1':
        c ="C"
    else:
        c=""
        
    if php =='1':
        php ="PHP"
    else:
        php=""    
        
    if java =='1':
        java ="Java"
    else:
        java=""   
        
    if py =='1':
        py ="Phython"
    else:
        py=""   
        
        
    file = open("StudentDetails.txt","a")
    file.write("\n\nFirst Name: "+First+"\n")
    file.write("Last Name: "+Last+"\n")
    file.write("Gender: "+gender+"\n")
    file.write("Phone Number: "+phone+"\n")
    file.write("Email: "+email+"\n")
    file.write("Address: "+address)
    file.write("Department: "+dep+" ")
    file.write("\nCourse: ")
    file.write(c+" ")
    file.write(php+" ")
    file.write(java+" ")
    file.write(py+" ")
   
    file.close()
    
    user = First + " has successfully Registered and Record is saved in StudentDetails.txt file"
    Label(win,text=user,fg="purple",bg="cyan",font = ("Calibri 12 bold")).place(x=12,y=560)
    print("Printing Data: ")
    print(First,Last,gender,phone,email,address,c,php,java,py)

def reset():
    Fname.set("")
    Lname.set("")
    Phonenum.set("")
    Email.set("")
    entry_Address.delete(1.0,END)




win =  Tk()

win.geometry("400x600")
win.configure(background="cyan")
win.title("Registration Form")

title = Label(win, text="Registration Form", bg="light green", width="300", height="2", fg="black", font = ("Calibri 24 bold  underline")).pack()

Fname = Label(win, text="First Name: ", bg="cyan", font=("Verdana 12 bold")).place(x=12, y=100)

Lname = Label(win, text="Last Name: ", bg="cyan", font=("Verdana 12 bold")).place(x=12, y=140)

Gender = Label(win, text="Gender: ",bg="cyan",font = ("Verdana 12 bold")).place(x=12,y=180)

Phonenum = Label(win, text="Phone Number: ",bg="cyan",font = ("Verdana 12 bold")).place(x=12,y=260)

email = Label(win, text="Email: ", bg="cyan", font=("Verdana 12 bold")).place(x=12, y=300)

Address = Label(win, text="Address: ",bg="cyan",font = ("Verdana 12 bold")).place(x=12,y=340)

radio = StringVar()
Male = Radiobutton(win, text="Male",bg="cyan",variable=radio,value="Male",font = ("Verdana 10 bold")).place(x=12,y=210)
Female = Radiobutton(win, text="Female",bg="cyan",variable=radio,value="Female",font = ("Verdana 10 bold")).place(x=120,y=210)

dept = Label(win,text="Department: ",bg="cyan",font = ("Verdana 12 bold")).place(x=12,y=380)
variable = StringVar(win)
variable.set("Dept: ")
w = OptionMenu(win, variable, "SE", "IT", "CS","PS")
w.place(x=140,y=380)


AC = Label(win,text="Courses you like to follow: ",bg="cyan",font = ("Verdana 12 bold")).place(x=12,y=420)
check = StringVar()
check2 = StringVar()
check3= StringVar()
check4 = StringVar()

C= Checkbutton(win, text = "C",bg="cyan",variable=check,font = ("Verdana 10 bold")).place(x=12,y=450)
PHP = Checkbutton(win, text = "PHP",variable=check2,bg="cyan",font = ("Verdana 10 bold")).place(x=12,y=480)
Java = Checkbutton(win, text = "Java",bg="cyan",variable=check3,font = ("Verdana 10 bold")).place(x=120,y=450)
Python = Checkbutton(win, text = "Python",bg="cyan",variable=check4,font = ("Verdana 10 bold")).place(x=120,y=480)

Fname = StringVar()
Lname = StringVar()
Gender = StringVar()
Phonenum=StringVar()
Email = StringVar()
Address  = StringVar()
Courses  = StringVar()

entry_Fname = Entry(win,textvariable = Fname,width=30)
entry_Fname.place(x=120,y=100)

entry_Lname = Entry(win,textvariable = Lname,width=30)
entry_Lname.place(x=120,y=140)

entry_Phonenum = Entry(win,textvariable = Phonenum,width=30)
entry_Phonenum.place(x=160,y=260)

entry_email = Entry(win,textvariable = Email,width=30)
entry_email.place(x=100,y=300)


entry_Address = Text(win,height=2,width=25)
entry_Address.place(x=120,y=340)

reset = Button(win, text="Reset", width="12",height="1",activebackground="red",command=reset, bg="light blue",font = ("Calibri 12 ")).place(x=24, y=520)
submit = Button(win, text="Submit", width="12",height="1",activebackground="violet", bg="Pink",command=filedreq,font = ("Calibri 12 ")).place(x=240, y=520)

win.mainloop()
```
